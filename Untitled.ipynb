```python
def draw_conformations(x, d, p, E, structs_per_row = 5):
    n_beads = x.shape[0]
    n_confs = x.shape[1]
    for i in range(n_confs):
        
        subplot(int(n_confs / structs_per_row)+1, structs_per_row, i+1)

        plot(x[:,i], c='k')

        for j_ind, j in enumerate(d[:,i]):
            if j == 1:
                scatter([j_ind],[x[j_ind,i]],marker=10,c='red')
            elif j == -1:
                scatter([j_ind],[x[j_ind,i]],marker=11,c='orange')

        for j_ind, j in enumerate(p[:,i]):
            if j != -1:
                if j_ind > j:
                    plot([j, j_ind],[x[j_ind,i],x[j_ind,i]],c='blue', linewidth=8)

        text(0,n_beads,E[i])
        
        xlim([-1,n_beads+1])
        ylim([-n_beads,n_beads])
        
        axhline(0,c='grey',linewidth=0.5)
        axis('off')
```

```python
def draw_traces(x,d,p,E, boltz_weight=False):
    n_beads = x.shape[0]
    n_confs = x.shape[1]
    
    min_E = np.min(E)
    for i in range(n_confs):
        if boltz_weight:
            wt = np.exp(-0.5*(E[i]+min_E))
        else:
            wt=0.2
        plot(x[:,i]+np.random.normal(scale=0.1,size=n_beads), linewidth=wt, c='k',alpha=0.5)
        for j_ind, j in enumerate(p[:,i]):
            if j != -1:
                if j_ind > j:
                    plot([j, j_ind],[x[j_ind,i],x[j_ind,i]],c='blue', linewidth=wt,alpha=0.5)

    axis('off')
```

```python
x, d, p, E = get_conformations('',sequence='UAUUA')
figure(figsize=(4,4))
draw_traces(x,d,p,E,boltz_weight=True)
```
